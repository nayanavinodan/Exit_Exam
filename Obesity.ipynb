```python
# removing outliers
out = data[['Age','Height','Weight','NCP']]
out
```

```python
for col in out.columns:
    q1 = out[col].quantile(0.25)
    q3 = out[col].quantile(0.75)
    iqr = q3 - q1

    # Define lower and upper bounds for outliers
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Clip the values to the defined bounds
    out[col] = out[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
plt.figure(figsize=(20,18))
sns.boxplot(out)
```

```python
# replace the data
data['Gender'] = data['Gender'].map({'Male':0,'Female':1})
data['family_history_with_overweight'] = data['family_history_with_overweight'].map({'yes':1,'no':0})
data['FAVC'] = data['FAVC'].map({'yes':1,'no':0})
data['CAEC'] = data['CAEC'].map({'Sometimes':1,'Frequently':2,'Always':3,'no':0})
data['SMOKE'] = data['SMOKE'].map({'yes':1,'no':0})
data['SCC'] = data['SCC'].map({'yes':1,'no':0})
data['CALC'] = data['CALC'].map({'Sometimes':1,'Frequently':2,'Always':3,'no':0})
data['MTRANS'] = data['MTRANS'].map({'Public_Transportation':0,'Walking':1,'Automobile':2,'Motorbike':3,'Bike':4})
```

```python
data.head()
```

```python
# encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['NObeyesdad']=le.fit_transform(data['NObeyesdad'])
```

```python
#x y split
x=data.drop('NObeyesdad',axis=1)
y=data['NObeyesdad']
```

```python
#standard scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit_transform(x)
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
#decission tree
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# hyperparameter tuning
# Define the parameter grid for RandomForestClassifier
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Initialize the RandomForestClassifier model
model = RandomForestClassifier(random_state=42)

# GridSearchCV for best hyperparameters
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, scoring='accuracy')
grid_search.fit(x_train, y_train)

# Best hyperparameters
print("Best Hyperparameters:", grid_search.best_params_)

# Evaluate the tuned model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Tuned Model Accuracy: {accuracy:.2f}')
```

```python
import pandas as pd

# Example new data as a DataFrame
new_data = pd.DataFrame({
    'Age': [21.0],  # Example values
    'Weight': [64.0],
    'family_history_with_overweight': [1],
    'NCP': [3.0],
    'CH2O': [	2.0],
    'FAF': [0.0],
    'CALC': [0],
    '	MTRANS': [0],

})

# Get the correct feature names from the scaler
original_features = scaler.get_feature_names_out()

new_data_reconstructed = pd.DataFrame(0, columns=original_features, index=new_data.index)


for feature in new_data.columns:
    if feature in original_features:
        new_data_reconstructed[feature] = new_data[feature]

new_data_scaled = pd.DataFrame(scaler.transform(new_data_reconstructed), columns=new_data_reconstructed.columns)
```

```python
# Make predictions using the loaded model
predictions = grid_search.predict(new_data_scaled)

# Print the predictions
print(predictions)
```
